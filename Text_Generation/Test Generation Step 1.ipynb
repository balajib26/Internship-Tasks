```python
# Generate Alphanumeric

import uuid

def my_random_string(string_length=10):
    """Returns a random string of length string_length."""

# Convert UUID format to a Python string.
    random = str(uuid.uuid4())

# Make all characters uppercase.
    random = random.upper()

# Remove the UUID '-'.
    random = random.replace("-","")

# Return the random string.
    return random[0:string_length]

print(my_random_string(10)) # For example, D9E50C
```

```python
# Concatenate list data
def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result
```

```python
# Generate Passport number
def generate_passport():
    lower_upper_alphabet = string.ascii_letters
    random_letter = random.choice(lower_upper_alphabet)
    random_letter= random_letter.upper()
    l=[]
    l.append(random_letter)
    for i in range(8):
        m=random.randint(0,9)
        l.append(m)
        g= concatenate_list_data(l) 
    return g
```

```python
# Generate alphanumeric of size 7
def id_generator(size=7, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
# Generate pan number
def generate_pan():
    q=[]
    for i in range(3):
        lower_upper_alphabet = string.ascii_letters
        random_letter = random.choice(lower_upper_alphabet)
        random_letter= random_letter.upper()
        q.append(random_letter)
    f= id_generator()
    q.append(f)
    r= concatenate_list_data(q) 
    return r
```

```python
# Generate aadhar card number
def generate_aadhar():
    l=[]
    p=random.randint(1,9)
    l.append(p)
    for i in range(11):
        m=random.randint(0,9)
        l.append(m)
        g= concatenate_list_data(l)
        
    return g
```

```python
# Transliteration to convert devanagari alphabets to english alphabets 
trn = Transliterator(source='hin', target='eng', build_lookup=True)
```

```python
# Generate 9999 samples of 8 data types(columns) name, address, phone, date of birth, pan, aadhar,
# drivers license and passport numbers to train the model.
for i in range(10000):
    name = faker_hindi.name()
    name_english= trn.transform(name)
    address = faker_hindi.address()
    address_english= trn.transform(address)
    phone = faker_hindi.phone_number()
    address2=' '.join(address.split())
    address_english2=' '.join(address_english.split())
    print(f'{name}, {address2}, {phone}')
    db=faker_hindi.date_of_birth()
    pan= generate_pan()
    aadhar= generate_aadhar()
    drive= my_random_string(13)
    passport= generate_passport()
    # name, address, date of birth, Aadhar Card Number, PAN Card number, Driving License number (improve)
    # passport
    a=[name_english,'\t', address_english2, '\t ' ,db.strftime("%Y-%m-%d"),'\t ' ,aadhar , '\t ', pan, '\t ', drive, '\t ', passport, '\n']
    file2 = open("faker2english.txt","a")
    file2.writelines(a)

file2.close()
```

```python
# Find location and city from address and store in address 2
def find_address2(address):
    func=re.compile(r'\d{6}')
    l=re.search(func,address).start()
    return address[:l]
```

```python
# Find pin code from address 
def find_pin_code(address):
    func=re.compile(r'\d{6}')
    l=re.search(func,address).start()
    return address[l:]
```
